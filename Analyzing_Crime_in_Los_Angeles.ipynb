```python
# Re-run this cell
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need

# Hour with highest frequency of crimes
# Se obtiene exclusivamente la hora a la que ocurrió el crimen, guardando este dato en una nueva columna llamada HOUR OCC
# Luego, se agrupa por esta nueva columna, obteniendo además el número de renglones en esa columna en el dataframe by_hour
crimes['HOUR OCC'] = pd.to_numeric(crimes['TIME OCC']) // 100
by_hour = crimes.groupby('HOUR OCC').size()

# Se obtiene la hora en la que ocurren mas crímenes y se guarda en la variable peak_crime_hour
hour = by_hour[by_hour == by_hour.max()].index
peak_crime_hour = int(hour[0])
```

```python
peak_crime_hour
```

Esto es, la hora a la que ocurren más crimenes es a las 12 del mediodía.

```python
# Area with largest frequency of night crimes
# Se obtiene un DataFrame night_crimes que tiene los crimenes cometidos entre las 10 de la noche y las 4 de la mañana
night_crimes = crimes[crimes['HOUR OCC'].isin([0,1,2,3,22,23])]
# Obtenemos night_crimes_by_area agrupando night_crimes por la columna AREA NAME, obteniendo además 
# el número de renglones (u observaciones) por área, y se obtiene el área con más crímenes nocturnos
night_crimes_by_area = night_crimes.groupby('AREA NAME').size()
area = night_crimes_by_area[night_crimes_by_area == night_crimes_by_area.max()].index
peak_night_crime_location = area[0]
```

```python
peak_night_crime_location
```

Así, el área donde ocurren más crimenes en la noche es denominada Central.

```python
# Crimes commited against victims by age group
# Creamos los intervalos de edades solicitados, así como las etiquetas.
bins = np.array([0, 17, 25, 34, 44, 54, 64, np.inf])
age_labels = ['<18','18-25', '26-34', '35-44', '45-54', '55-64', '65+']
# Contamos el número de observaciones por grupo de edad.
crimes['Age group'] = pd.cut(crimes['Vict Age'], bins = bins, labels = age_labels)
victim_ages = crimes['Age group'].value_counts()
```
